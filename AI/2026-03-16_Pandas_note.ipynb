## 1. Pandas 개념

```python
# Pandas
import pandas as pd
```

## 2. Series


```python
# -------------------------------------------------
# Series
#
# Series는 Pandas의 1차원 데이터 구조이다.
#
# NumPy 배열과 비슷하지만
# "index"라는 개념이 추가되어 있다.
#
# 구조
#
# index  value
# 0      10
# 1      20
# 2      30
#
# 즉
# index + value 형태의 데이터 구조이다.
# -------------------------------------------------


# -------------------------------------------------
# Series 생성

s = pd.Series([10, 20, 30, 40])

print(s)

# 결과
#
# 0    10
# 1    20
# 2    30
# 3    40
# dtype: int64

# 왼쪽 → index
# 오른쪽 → 실제 데이터 값


# -------------------------------------------------
# index 확인

print(s.index)

# RangeIndex(start=0, stop=4, step=1)

# 기본 index는 0부터 시작한다.


# -------------------------------------------------
# values 확인

print(s.values)

# array([10, 20, 30, 40])


# -------------------------------------------------
# index 직접 지정

s = pd.Series([10, 20, 30], index=["a", "b", "c"])

print(s)

# 결과
#
# a    10
# b    20
# c    30
# dtype: int64


# -------------------------------------------------
# 특정 값 선택 (인덱싱)

print(s["a"])   # 10

# index를 이용해서 값을 선택할 수 있다.


# -------------------------------------------------
# 슬라이싱

print(s["a":"b"])

# 결과
#
# a    10
# b    20

# Pandas 슬라이싱은
# 끝 index도 포함된다.


# -------------------------------------------------
# 조건 선택 (Boolean indexing)

s = pd.Series([10, 20, 30, 40, 50])

print(s[s > 30])

# 결과
#
# 3    40
# 4    50

# 조건을 만족하는 데이터만 선택된다.


# -------------------------------------------------
# Series 연산

print(s + 10)

# 결과
# 20
# 30
# 40
# 50
# 60

# 모든 값에 10이 더해진다.


# -------------------------------------------------
# 통계 함수

print(s.sum())   # 합
print(s.mean())  # 평균
print(s.max())   # 최대값
print(s.min())   # 최소값


# -------------------------------------------------
# 정리
#
# Series 특징
#
# 1) 1차원 데이터 구조
# 2) index + value 구조
# 3) NumPy 배열과 유사하지만 index 개념 존재
# 4) 인덱싱 / 슬라이싱 / 조건 선택 가능
#
# 대부분의 데이터 분석에서는
# Series 여러 개가 모여 DataFrame을 만든다.
# -------------------------------------------------
```

## 3. DataFrame

```python
# -------------------------------------------------
# DataFrame 생성 방법
#
# Pandas DataFrame은 다양한 방식으로 생성할 수 있다.
#
# 대표적인 방법
#
# 1 dictionary
# 2 list
# 3 list of dict
# 4 numpy array
# 5 series
# 6 csv 파일
# -------------------------------------------------


# -------------------------------------------------
# 1 dictionary로 DataFrame 생성
#
# 가장 많이 사용하는 방법이다.

data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "score": [90, 85, 95]
}

df = pd.DataFrame(data)

print(df)


# -------------------------------------------------
# 2 list로 DataFrame 생성

data = [
    ["Alice", 25, 90],
    ["Bob", 30, 85],
    ["Charlie", 35, 95]
]

df = pd.DataFrame(data, columns=["name", "age", "score"])

print(df)


# -------------------------------------------------
# 3 list of dictionary
#
# JSON 데이터 구조와 매우 비슷하다.

data = [
    {"name": "Alice", "age": 25, "score": 90},
    {"name": "Bob", "age": 30, "score": 85},
    {"name": "Charlie", "age": 35, "score": 95}
]

df = pd.DataFrame(data)

print(df)


# -------------------------------------------------
# 4 numpy 배열로 DataFrame 생성

arr = np.array([
    ["Alice", 25, 90],
    ["Bob", 30, 85],
    ["Charlie", 35, 95]
])

df = pd.DataFrame(arr, columns=["name", "age", "score"])

print(df)


# -------------------------------------------------
# 5 Series로 DataFrame 생성

name = pd.Series(["Alice", "Bob", "Charlie"])
age = pd.Series([25, 30, 35])
score = pd.Series([90, 85, 95])

df = pd.DataFrame({
    "name": name,
    "age": age,
    "score": score
})

print(df)


# -------------------------------------------------
# 6 csv 파일에서 DataFrame 생성
#
# 실제 데이터 분석에서 가장 많이 사용하는 방식이다.
#
# CSV = Comma Separated Values

# df = pd.read_csv("data.csv")

# print(df)


# -------------------------------------------------
# 정리
#
# DataFrame 생성 방법
#
# dictionary      → 가장 많이 사용
# list            → 구조가 단순할 때
# list of dict    → JSON 데이터
# numpy array     → NumPy 데이터 변환
# series          → Series 결합
# csv             → 실제 데이터 분석
# -------------------------------------------------
```

## 4. 데이터 선택 (column / loc / iloc)


```python
# -------------------------------------------------
# 데이터 선택 (Data Selection)
#
# Pandas에서 데이터를 선택하는 방법은 크게 3가지이다.
#
# 1 column 선택
# 2 loc (label 기반 선택)
# 3 iloc (index 기반 선택)
#
# 이 세 가지는 Pandas 데이터 처리의 핵심이다.
# -------------------------------------------------


# -------------------------------------------------
# 예제 DataFrame 생성

data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "score": [90, 85, 95]
}

df = pd.DataFrame(data)

print(df)

# 결과
#
#       name  age  score
# 0    Alice   25     90
# 1      Bob   30     85
# 2  Charlie   35     95


# =================================================
# 1 column 선택
# =================================================

print(df["name"])

# 결과
#
# 0      Alice
# 1        Bob
# 2    Charlie
#
# dtype: object

# 하나의 column을 선택하면
# 결과는 Series 형태가 된다.


# -------------------------------------------------
# 여러 column 선택

print(df[["name", "score"]])

# 결과
#
#       name  score
# 0    Alice     90
# 1      Bob     85
# 2  Charlie     95

# 여러 column 선택 시
# 반드시 리스트 형태로 지정해야 한다.


# =================================================
# 2 row 선택
# =================================================

print(df.loc[0])

# 결과
#
# name     Alice
# age         25
# score       90


# -------------------------------------------------
# 여러 row 선택

print(df.loc[[0, 2]])

# 결과
#
#       name  age  score
# 0    Alice   25     90
# 2  Charlie   35     95


# =================================================
# 3 loc (label 기반 선택)
# =================================================
#
# loc은 index 이름(label)을 기준으로 데이터를 선택한다.

print(df.loc[0, "name"])

# 결과
# Alice


# -------------------------------------------------
# 특정 row와 column 선택

print(df.loc[1, "score"])

# 결과
# 85


# -------------------------------------------------
# 여러 row와 column 선택

print(df.loc[[0, 2], ["name", "score"]])

# 결과
#
#       name  score
# 0    Alice     90
# 2  Charlie     95


# =================================================
# 4 iloc (index 기반 선택)
# =================================================
#
# iloc은 숫자 index를 기준으로 데이터를 선택한다.

print(df.iloc[0])

# 결과
#
# name     Alice
# age         25
# score       90


# -------------------------------------------------
# 특정 위치 값 선택

print(df.iloc[0, 1])

# 결과
# 25


# -------------------------------------------------
# 여러 행과 열 선택

print(df.iloc[[0,2], [0,2]])

# 결과
#
#       name  score
# 0    Alice     90
# 2  Charlie     95


# =================================================
# loc vs iloc 정리
# =================================================
#
# loc
# → label 기준 선택
#
# iloc
# → 숫자 index 기준 선택
#
# 예
#
# df.loc[0, "name"]
#
# df.iloc[0, 1]
#
# -------------------------------------------------
```

### 5. 조건 필터링

```python
# -------------------------------------------------
# 조건 필터링 (Boolean Filtering)
#
# 조건을 이용해서 원하는 데이터만 선택하는 방법이다.
#
# 형식
#
# df[조건]
#
# 조건이 True인 행만 선택된다.
# -------------------------------------------------


# -------------------------------------------------
# 예제 DataFrame 생성

data = {
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [25, 30, 35, 28],
    "score": [90, 85, 95, 70]
}

df = pd.DataFrame(data)

print(df)

# 결과
#
#       name  age  score
# 0    Alice   25     90
# 1      Bob   30     85
# 2  Charlie   35     95
# 3    David   28     70


# -------------------------------------------------
# 조건 확인

print(df["age"] > 30)

# 결과
#
# 0    False
# 1    False
# 2     True
# 3    False

# 각 행이 조건을 만족하는지 True/False로 표시된다.


# -------------------------------------------------
# 조건 필터링

print(df[df["age"] > 30])

# 결과
#
#       name  age  score
# 2  Charlie   35     95

# age가 30보다 큰 행만 선택된다.


# -------------------------------------------------
# 다른 조건 예시

print(df[df["score"] >= 90])

# 결과
#
#       name  age  score
# 0    Alice   25     90
# 2  Charlie   35     95


# -------------------------------------------------
# 여러 조건 사용 (&)
#
# 조건1 AND 조건2

print(df[(df["age"] > 25) & (df["score"] > 80)])

# 결과
#
#       name  age  score
# 1      Bob   30     85
# 2  Charlie   35     95


# -------------------------------------------------
# 여러 조건 사용 (|)
#
# 조건1 OR 조건2

print(df[(df["age"] < 28) | (df["score"] > 90)])

# 결과
#
#       name  age  score
# 0    Alice   25     90
# 2  Charlie   35     95


# -------------------------------------------------
# 특정 column만 출력

print(df[df["score"] > 80]["name"])

# 결과
#
# 0      Alice
# 1        Bob
# 2    Charlie


# -------------------------------------------------
# loc와 함께 사용

print(df.loc[df["score"] > 80, ["name", "score"]])

# 결과
#
#       name  score
# 0    Alice     90
# 1      Bob     85
# 2  Charlie     95


# -------------------------------------------------
# 정리
#
# 조건 필터링 형식
#
# df[조건]
#
# 예
#
# df[df["age"] > 30]
#
#
# 여러 조건
#
# AND → &
# OR  → |
#
# 예
#
# df[(조건1) & (조건2)]
#
# 조건 필터링은
# 데이터 분석에서 가장 많이 사용하는 기능이다.
# -------------------------------------------------
```

### 6. 정렬

```python
# -------------------------------------------------
# 정렬 (Sorting)
#
# Pandas에서는 데이터를 정렬할 때
# sort_values() 함수를 사용한다.
#
# 형식
#
# df.sort_values(by="column")
#
# 특정 column 값을 기준으로 정렬한다.
# -------------------------------------------------


# -------------------------------------------------
# 예제 DataFrame 생성

data = {
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [25, 30, 35, 28],
    "score": [90, 85, 95, 70]
}

df = pd.DataFrame(data)

print(df)


# =================================================
# 1 column 기준 정렬
# =================================================

print(df.sort_values(by="age"))

# age 기준 오름차순 정렬

# 결과
#
#       name  age  score
# 0    Alice   25     90
# 3    David   28     70
# 1      Bob   30     85
# 2  Charlie   35     95


# -------------------------------------------------
# score 기준 정렬

print(df.sort_values(by="score"))


# =================================================
# 내림차순 정렬
# =================================================

print(df.sort_values(by="score", ascending=False))

# ascending=False
# → 내림차순 정렬


# 결과
#
#       name  age  score
# 2  Charlie   35     95
# 0    Alice   25     90
# 1      Bob   30     85
# 3    David   28     70


# =================================================
# 여러 column 기준 정렬
# =================================================

print(df.sort_values(by=["age", "score"]))

# 먼저 age로 정렬
# 같은 age일 경우 score 기준으로 정렬


# -------------------------------------------------
# 여러 column + 정렬 방향

print(df.sort_values(by=["age", "score"], ascending=[True, False]))

# age → 오름차순
# score → 내림차순


# =================================================
# index 기준 정렬
# =================================================

print(df.sort_index())

# index 순서대로 정렬


# -------------------------------------------------
# index 내림차순

print(df.sort_index(ascending=False))


# -------------------------------------------------
# 정렬 후 index 초기화

print(df.sort_values(by="score").reset_index(drop=True))

# reset_index()
# → index 다시 정리


# -------------------------------------------------
# 정리
#
# sort_values()
# → column 기준 정렬
#
# sort_index()
# → index 기준 정렬
#
# ascending=False
# → 내림차순
#
# reset_index()
# → 정렬 후 index 초기화
# -------------------------------------------------
```
